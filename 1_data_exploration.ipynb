```python
from datasets import load_dataset

print("Starting load...")
dataset = load_dataset("birgermoell/icd10-clinical-notes")
print("Loaded successfully!")
print(dataset)
```

```python
import pandas as pd

train_df = dataset["train"].to_pandas()
test_df = dataset["test"].to_pandas()

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("Columns:", train_df.columns.tolist())
train_df.head()
```

```python
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

print(train_df[["code", "language", "journal_note"]].head(5))
```

```python
diabetes_mask_train = (
    train_df["code"].astype(str).str.startswith("E10") |
    train_df["code"].astype(str).str.startswith("E11")
)

diabetes_mask_test = (
    test_df["code"].astype(str).str.startswith("E10") |
    test_df["code"].astype(str).str.startswith("E11")
)

train_df["diabetes_label"] = diabetes_mask_train.astype(int)
test_df["diabetes_label"] = diabetes_mask_test.astype(int)

print("Train label counts:")
print(train_df["diabetes_label"].value_counts())

print("\nTest label counts:")
print(test_df["diabetes_label"].value_counts())
```
